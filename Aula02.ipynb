```python
# apagar coluna "registro_sistema"
dados_tratados1 = emprestimos_completo.drop("registro_sistema", axis=1)
dados_tratados1.info()
```

```python
# convertendo a coluna  (“matricula_ou_siape”) em string
dados_tratados1['matricula_ou_siape'] = dados_tratados1['matricula_ou_siape'].apply(lambda x: f'{x:.2f}')
dados_tratados1.info()
```

```python
# nova coluna CDU, com base na numera "localização"ArithmeticError

tabela_para_teste = dados_tratados1.head(100)
tabela_para_teste
```

```python
# funcao para classificação da coluna

def categorizacao(localizacao):
    if 0 <= localizacao <= 99:
        return "Generalidades. Ciência e conhecimento."
    elif 100 <= localizacao <= 199:
        return "Filosofia e psicologia."
    elif 200 <= localizacao <= 299:
        return "Religião."
    elif 300 <= localizacao <= 399:
        return "Ciências sociais."
    elif 400 <= localizacao <= 499:
        return "Classe vaga. Provisoriamente não ocupada."
    elif 500 <= localizacao <= 599:
        return "Matemática e ciências naturais."
    elif 600 <= localizacao <= 699:
        return "Ciências aplicadas."
    elif 700 <= localizacao <= 799:
        return "Belas artes."
    elif 800 <= localizacao <= 899:
        return "Linguagem. Língua. Linguística."
    elif 900 <= localizacao <= 999:
        return "Geografia. Biografia. História."
    else:
        return "SEM CATEGORIA DEFINIDA"
    
categorizacao(1000)   

```

```python
tabela_para_teste["CDU"] = tabela_para_teste["localizacao"].apply(categorizacao)
tabela_para_teste
```

```python
tabela_aula02 = dados_tratados1

tabela_aula02["CDU"] = tabela_aula02["localizacao"].apply(categorizacao)

tabela_aula02
```
